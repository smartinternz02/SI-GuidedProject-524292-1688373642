### Find the outliers and replace the outliers

```python
import pandas as pd
import numpy as np

# Identify outliers using z-score
num = ['area','bedrooms','bathrooms','stories','parking']  # Select numeric columns to check for outliers
threshold = 2  # Set the threshold for z-score

for c in num:
    z_scores = (df[c] - df[c].mean()) / df[c].std()
    outlier_indices = np.abs(z_scores) > threshold
    outliers = df[outlier_indices]
    print("Outliers in", c, ":", outliers)

# Replace outliers with NaN or a specific value
df['area'] = np.where(np.abs(z_scores) > threshold, np.nan, df['area'])
df['bedrooms'] = np.where(np.abs(z_scores) > threshold, np.nan, df['bedrooms'])
df['bathrooms'] = np.where(np.abs(z_scores) > threshold, np.nan, df['bathrooms'])
df['stories'] = np.where(np.abs(z_scores) > threshold, np.nan, df['stories'])
df['parking'] = np.where(np.abs(z_scores) > threshold, np.nan, df['parking'])

# Handle outliers based on specific requirements (e.g., replace with mean, median, etc.)
df['area'].fillna(df['area'].median(), inplace=True)
df['bedrooms'].fillna(df['bedrooms'].mean(), inplace=True)
df['bathrooms'].fillna(df['bathrooms'].mean(), inplace=True)
df['stories'].fillna(df['stories'].mean(), inplace=True)
df['parking'].fillna(df['parking'].mean(), inplace=True)
```

### Check for categorical values and perform encoding

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
df['mainroad']=l.fit_transform(df['mainroad'])
df['guestroom']=l.fit_transform(df['guestroom'])
df['basement']=l.fit_transform(df['basement'])
df['hotwaterheating']=l.fit_transform(df['hotwaterheating'])
df['airconditioning']=l.fit_transform(df['airconditioning'])
df['furnishingstatus']=l.fit_transform(df['furnishingstatus'])
```

### Split the data

```python
X=df.drop('price',axis=1)
y=df['price']
```

### Scale the independent variables

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled=pd.DataFrame(X_scaled,columns=X.columns)
```

### 10 Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=0)
```
