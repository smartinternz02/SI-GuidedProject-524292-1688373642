### find the outliers and replace them

```python
import pandas as pd
import numpy as np

# Identify outliers using z-score
num = ['Age', 'Fare']  # Select numeric columns to check for outliers
threshold = 2  # Set the threshold for z-score

for c in num:
    z_scores = (df[c] - df[c].mean()) / df[c].std()
    outlier_indices = np.abs(z_scores) > threshold
    outliers = df[outlier_indices]
    print("Outliers in", c, ":", outliers)

# Replace outliers with NaN or a specific value
df['Age'] = np.where(np.abs(z_scores) > threshold, np.nan, df['Age'])
df['Fare'] = np.where(np.abs(z_scores) > threshold, np.nan, df['Fare'])

# Handle outliers based on specific requirements (e.g., replace with mean, median, etc.)
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
```
